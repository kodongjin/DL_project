# 이미지 합성

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os

# 멜 스펙트로그램 설정
n_fft = 2048       # FFT 컴포넌트의 수
hop_length = 512   # 연속적인 프레임 간 샘플 수
n_mels = 128       # 멜 밴드의 수
```

**Background sound**

```python
import os

# 백그라운드 파일이 들어있는 디렉터리 경로
directory_path = '/content/drive/MyDrive/1DL_data_p/background/'

# 각 백그라운드 디렉터리에 대한 파일 이름 목록을 저장할 리스트들
background_lists = []

# 백그라운드 디렉터리들에 대해 반복
for i in range(1, 7):
    # 백그라운드 디렉터리의 경로
    dir_path = os.path.join(directory_path + 'back' + str(i))
    # 백그라운드 디렉터리 안에 있는 파일 이름 목록 가져오기
    file_names = os.listdir(dir_path)
    # 해당 리스트에 파일 이름 목록 추가
    background_lists.append(file_names)

# 결과 확인
for i, file_list in enumerate(background_lists, 1):
    print(f"background_list{i}: {file_list}")
```

```python
import os

# 백그라운드 파일이 들어있는 디렉터리 경로
directory_path = '/content/drive/MyDrive/1DL_data_p/event/'

# 각 백그라운드 디렉터리에 대한 파일 이름 목록을 저장할 리스트들
event_lists = []

# 백그라운드 디렉터리들에 대해 반복
for i in range(1, 4):
    # 백그라운드 디렉터리의 경로
    dir_path = os.path.join(directory_path + 'event' + str(i))
    # 백그라운드 디렉터리 안에 있는 파일 이름 목록 가져오기
    file_names = os.listdir(dir_path)
    # 해당 리스트에 파일 이름 목록 추가
    event_lists.append(file_names)

# 결과 확인
for i, file_list in enumerate(event_lists, 1):
    print(f"event_list{i}: {file_list}")
```

# 음성파일합성 및 이미지생성

**background_event(a,b)**
- a번째 background 파일 / b번째 event 파일 합성하여 -> 이미지파일 저장
- 저장 위치 : 1DL_data_p/mel_spectrogram_image/image_b***a***_e***b***

```python
random.randint(1,5)
```

```python
def background_event(a,b):
  # a번째 background 파일 / b번째 event 파일 합성하여 -> 이미지파일 저장

  # background 음성 파일 불러오기
  for i in range(len(background_lists[a-1])):
    file = "/content/drive/MyDrive/1DL_data_p/background/back" + str(a) + "/" + background_lists[a-1][i]
    b_data, b_sample_rate = librosa.load(file, sr = 48000)

    # 15초 보다 작은 데이터는 삭제
    if len(b_data) / b_sample_rate < 15.0 :
      continue

  # event 음성 파일 불러오기
    for j in range(len(event_lists[b-1])):
      file = "/content/drive/MyDrive/1DL_data_p/event/event" + str(b) + "/" + event_lists[b-1][j]
      e_data, e_sample_rate = librosa.load(file, sr = 48000)


      # 데이터 합성 (background + event)
      # 1. 랜덤한 위치에 / 2. 랜덤한 크기 차이
      start_point = random.randint(0,15 * 48000)

      zero_list = np.zeros(start_point)
      zero_long_list = np.zeros(15 * 48000)
      e_data_1 = np.concatenate((zero_list, np.array(e_data), np.array(zero_long_list)))

      e_data_1 = e_data_1[:15*48000]

      data_systehsis = (1-random.randint(1,5)) * b_data + random.randint(1,5) * e_data_1

      # 멜 스펙트로그램 계산
      S = librosa.feature.melspectrogram(y=data_systehsis, sr=b_sample_rate, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)

      # 파워 스펙트로그램을 dB 단위로 변환 (시각화를 위해)
      S_dB = librosa.power_to_db(S, ref=np.max)

      fig, ax = plt.subplots(figsize=(10, 4))
      ax.set_position([0, 0, 1, 1]) # (모서리위치, 가로 길이, 세로 길이)
      ax.set_axis_off() # 객체 축 제거

      librosa.display.specshow(S_dB, sr=b_sample_rate, hop_length=hop_length, x_axis='time', y_axis='mel',ax=ax)
                              # (db단위로변환멜스펙, 샘플링주파수, fft창연속된간격, x축시간, y축멜주파수, 이전에생성한axes에생성)
      plt.savefig('/content/drive/MyDrive/1DL_data_p/mel_spectrogram_image/' + "image_e" + str(b) +'/' + "img_b" + str(a) + "_" + str(i) + "_" + str(j) + ".png")
```

```python
# background_event(background_num, event_num)
event_num = 1

background_event(1, event_num)
background_event(2, event_num)
background_event(3, event_num)
background_event(4, event_num)
background_event(5, event_num)
background_event(6, event_num)
```

# 이미지를 -> .npy 데이터로

```python
b = 1 # event 번호

image_list = os.listdir('/content/drive/MyDrive/1DL_data_p/mel_spectrogram_image/image_e'+ str(b))
image_list
```

```python
from PIL import Image

def imgTo_npy(b) :
  x = []
  for i in range(len(image_list)):
    # 이미지 파일 열기
    image = Image.open('/content/drive/MyDrive/1DL_data_p/mel_spectrogram_image/image_e'+ str(b) + "/" + image_list[i])
    image_array = np.array(image)
    x_data_1 = image_array[:,:,0:3]

    x.append(x_data_1)

  x_data = np.array(x)

  np.save("/content/drive/MyDrive/1DL_data_p/x_data/data_e" + str(b) + ".npy", x_data)
```

```python
event_num = 1

imgTo_npy(event_num)
```
